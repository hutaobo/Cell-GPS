```python
# Panel mapping:
# - Supplementary Figure 11a-n, shared setup. Imports plotting and table libraries used by unclustered lung fibrosis heatmaps across samples.
#

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
sns.set_theme(context="talk", style="white")
```

```python
# Panel mapping:
# - Supplementary Figure 11a-n. Iterates through lung fibrosis Xenium samples, applies custom cell groups, and writes one unclustered SSS heatmap per sample.
#

from sfplot import load_xenium_data, compute_cophenetic_distances_from_adata, plot_cophenetic_heatmap

BASE_DIR = Path("Y:/long/publication_datasets/Vannan_2023_Lung_Fibrosis/Xenium")
ANNOTATION_DIR = BASE_DIR / "custom cell groups"
OUTPUT_DIR = Path("outputs/supp_fig_11_lung_fibrosis_unclustered_heatmaps")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

for folder in [p for p in BASE_DIR.iterdir() if p.is_dir()]:
    sample = folder.name
    annotation_file = ANNOTATION_DIR / f"{sample}.csv"
    if not annotation_file.exists():
        continue
    adata = load_xenium_data(str(folder))
    annotation = pd.read_csv(annotation_file, index_col=0)
    adata.obs["group"] = adata.obs["cell_id"].map(annotation["group"])
    row_coph, col_coph = compute_cophenetic_distances_from_adata(adata, cluster_col="group")
    row_coph.to_csv(OUTPUT_DIR / f"{sample}_row_coph.csv")
    plot_cophenetic_heatmap(row_coph, matrix_name="row_coph", output_filename=str(OUTPUT_DIR / f"{sample}_row_coph.pdf"), sample=sample)
```
