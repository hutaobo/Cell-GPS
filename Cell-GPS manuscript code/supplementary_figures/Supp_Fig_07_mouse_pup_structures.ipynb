```python
# Panel mapping:
# - Supplementary Figure 7a-d, shared setup. Imports plotting and table libraries used by the hierarchical mouse pup structure panels.
#

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
sns.set_theme(context="talk", style="white")
```

```python
# Panel mapping:
# - Supplementary Figure 7a-d. Loads mouse pup Xenium data and writes spatial panels for Cluster, k_2, k_3, and k_6 structure annotations.
#

import squidpy as sq
from sfplot import load_xenium_data

XENIUM_DIR = Path("Y:/long/10X_datasets/Xenium/Xenium_5K/Xenium_Prime_Mouse_Pup_FFPE_outs")
OUTPUT_DIR = Path("outputs/supp_fig_07_mouse_pup_structures")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
adata = load_xenium_data(str(XENIUM_DIR), normalize=False)

for key in ["Cluster", "k_2", "k_3", "k_6"]:
    if key not in adata.obs:
        print(f"Skip {key}: column not present in adata.obs")
        continue
    sq.pl.spatial_scatter(adata, color=key, shape=None, img=False, figsize=(8, 12), show=False)
    plt.gcf().savefig(OUTPUT_DIR / f"mouse_pup_{key}.pdf", bbox_inches="tight", dpi=300)
    plt.close()
```
