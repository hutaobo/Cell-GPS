```python
# Panel mapping:
# - Supplementary Figure 9a-g, shared setup. Imports plotting and table libraries used by the Squidpy parameter-sensitivity panels.
#

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
sns.set_theme(context="talk", style="white")
```

```python
# Panel mapping:
# - Supplementary Figure 9a-g. Re-runs Squidpy neighborhood enrichment across kNN and radius settings and exports one heatmap per parameter setting.
#

import squidpy as sq
from sfplot import load_xenium_data

XENIUM_DIR = Path("Y:/long/10X_datasets/Xenium/Xenium_5K/Xenium_Prime_Mouse_Pup_FFPE_outs")
OUTPUT_DIR = Path("outputs/supp_fig_09_squidpy_parameter_sensitivity")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
adata = load_xenium_data(str(XENIUM_DIR), normalize=False)
cluster_key = "annotation" if "annotation" in adata.obs else "Cluster"
adata.obs[cluster_key] = adata.obs[cluster_key].astype("category")

for n_neighs in [4, 6, 8, 10, 15]:
    tmp = adata.copy()
    sq.gr.spatial_neighbors(tmp, coord_type="generic", n_neighs=n_neighs)
    sq.gr.nhood_enrichment(tmp, cluster_key=cluster_key)
    sq.pl.nhood_enrichment(tmp, cluster_key=cluster_key, figsize=(9, 9), method="average", show=False)
    plt.gcf().savefig(OUTPUT_DIR / f"squidpy_nhood_enrichment_n{n_neighs}.pdf", bbox_inches="tight")
    plt.close()

for radius in [50, 100]:
    tmp = adata.copy()
    sq.gr.spatial_neighbors(tmp, coord_type="generic", radius=radius)
    sq.gr.nhood_enrichment(tmp, cluster_key=cluster_key)
    sq.pl.nhood_enrichment(tmp, cluster_key=cluster_key, figsize=(9, 9), method="average", show=False)
    plt.gcf().savefig(OUTPUT_DIR / f"squidpy_nhood_enrichment_r{radius}.pdf", bbox_inches="tight")
    plt.close()
```
