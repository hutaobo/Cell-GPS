```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
sns.set_theme(context="talk", style="white")
```

```python
LANDMARK_DIR = Path("D:/GitHub/sfplot/sfplot-manuscript/t_by_c_SSc_2_1_1_Landmark")
A100_COPY = Path("/data/taobo.hu/projects/sfplot-manuscript/t_by_c_SSc_2_1_1_Landmark")
OUTPUT_DIR = Path("outputs/supp_table_3_SSc_landmark_transcript_SSS")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
source_dir = LANDMARK_DIR if LANDMARK_DIR.exists() else A100_COPY

result = pd.read_csv(source_dir / "t_and_c_result_SSc_2_1_1.csv", index_col=0)
counts = pd.read_csv(source_dir / "GeneCounts_byCluster_SSc_2_1_1.csv", index_col=0)
segment_free = pd.read_csv(source_dir / "StructureMap_table_SSc_2_1_1_Landmark_Segment_free.csv", index_col=0)

table = result.stack(dropna=True).rename("SSS").reset_index()
table.columns = ["gene", "landmark_cluster", "SSS"]
counts_long = counts.stack(dropna=True).rename("transcript_count").reset_index()
counts_long.columns = ["gene", "landmark_cluster", "transcript_count"]
table = table.merge(counts_long, on=["gene", "landmark_cluster"], how="left")
table.to_csv(OUTPUT_DIR / "supplementary_table_3_SSc_landmark_transcript_SSS.csv", index=False)
segment_free.to_csv(OUTPUT_DIR / "SSc_landmark_segment_free_StructureMap.csv")
table.head()
```
