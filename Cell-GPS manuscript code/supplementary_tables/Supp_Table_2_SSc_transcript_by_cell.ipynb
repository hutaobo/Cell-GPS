```python
# Panel mapping:
# - Supplementary Table 2, shared setup. Imports table and plotting libraries used to reshape SSc transcript-by-cell results.
#

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
sns.set_theme(context="talk", style="white")
```

```python
# Panel mapping:
# - Supplementary Table 2. Loads raw SSc transcript-by-cell SSS and count matrices, reshapes them to long format, and writes the manuscript table source.
#

RAW_TBC_DIR = Path("/data/taobo.hu/projects/ILD/t_by_c_SSc_2_1_1_raw")
LOCAL_COPY = Path("C:/Users/taobo.hu/Downloads/t_by_c_SSc_2_1_1_raw")
OUTPUT_DIR = Path("outputs/supp_table_2_SSc_transcript_by_cell")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
source_dir = RAW_TBC_DIR if RAW_TBC_DIR.exists() else LOCAL_COPY

result = pd.read_csv(source_dir / "t_and_c_result_SSc_2_1_1_raw.csv", index_col=0)
counts = pd.read_csv(source_dir / "GeneCounts_byCluster_SSc_2_1_1_raw.csv", index_col=0)
table = result.stack(dropna=True).rename("SSS").reset_index()
table.columns = ["gene", "cell_group", "SSS"]
counts_long = counts.stack(dropna=True).rename("transcript_count").reset_index()
counts_long.columns = ["gene", "cell_group", "transcript_count"]
table = table.merge(counts_long, on=["gene", "cell_group"], how="left")
table.to_csv(OUTPUT_DIR / "supplementary_table_2_SSc_transcript_by_cell_SSS.csv", index=False)
table.head()
```
